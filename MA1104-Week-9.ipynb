# Potential Function



$f(x,y) = 5(1-x+ 0.01x^{5}+y^{3})e^{-x^{2}-y^{2}}$

```python
def f(x, y):
    return 5*(1 - x + 0.01*x ** 5 + y ** 3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y)

pl.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap='jet')
C = pl.contour(X, Y, f(X, Y), 8, colors='black', linewidth=.5)
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = pl.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
Z = f(X,Y)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
```
