# <font face="sans-serif"> Calibration of the Heston model

<font face="sans-serif" size=3>
    
In this notebook, we present the calibration of the Heston model. The pricing is implemented using the Carr-Madan FFT, which can be found in the <a href=hestonpricer.py>hestonpricer</a> module. The <a href=blackscholespricer.py>blackscholespricer</a> module contains the pricing under the usual Black-Scholes model.

```python
import time
import numpy as np
import pandas as pd
from  pandas.io.data import *
import hestonpricer as heston
import blackscholespricer as bs
from scipy.stats import norm
from scipy import optimize
```

# <font face="sans-serif"> Market data and the calibration function

<font face="sans-serif", size=3>

We first define two important functions: 


1. heston_calibration(optiondata): This returns the model parameters and model prices.

2. market_price(optiondata): This computes the market option prices from implied volatility.

Both functions require the market (option) data as input. The csv file is attached here: 

<a href="marketdata.csv">marketdata.csv</a>

```python
# read and display the option data file 

mydata = pd.read_csv('marketdata.csv')
mydata
```

```python
def heston_calibration(optiondata):
    """ calibrate heston parameters [lambd, rho, eta, v0, v_bar] 
    using market option implied vol, and compute model price """
    
    N = len(optiondata)
    S = optiondata['Spot']
    r = optiondata['iRate']
    q = optiondata['dRate']
    K = optiondata['Strike']
    market_vol = optiondata['IVMid']
    T = np.zeros(N)
    
    for i in range(0,N):
        ndays = (pd.to_datetime(optiondata['dtExpiry'].ix[i])-pd.to_datetime(optiondata['dtTrade'].ix[i])).days
        T[i] = ndays/360 
        
    market_price = np.zeros(N)
    
    for i in range(0,N):
        market_price[i] = bs.vanilla_price(S[i], K[i], r[i]-q[i], market_vol[i], T[i], 'C')
        
    model_price = np.zeros(N)
 
    # objective function is the root mean squared error of vol
    sq_err = np.zeros(N) # sum of (market price - model price)**2
    def objfunc(params):
        lambd, rho, eta, v0, v_bar = params
        for i in range(0,N):
            model_price[i] = heston.callprice(S[i],K[i],r[i]-q[i],T[i],lambd,rho,eta,v0,v_bar)
            sq_err[i] = (market_price[i]-model_price[i])**2
        return sqrt((1/N)*sq_err.sum())
    
    # initial guess of the parameters
    ival=(1,-0.5,0.5, 0.05, 0.01) 
    
    # bounds on each parameter
    bnd = ((0,10), (-1,1), (0,1), (0,1), (0,1))
    
    # using optimize.minimize to find optimal values of paramters
    # other methods include 'Nelder-Mead', 'L-BFGS-B'
    output = optimize.minimize(objfunc, ival, method='SLSQP', bounds=bnd, options={'maxiter':10000})
    
    return output, model_price
```

```python
def market_price(optiondata):
    """compute option prices from implied volatility"""
    N = len(optiondata)
    S = optiondata['Spot']
    r = optiondata['iRate']
    q = optiondata['dRate']
    K = optiondata['Strike']
    market_vol = optiondata['IVMid']
    T = np.zeros(N)
    
    for i in range(0,N):
        ndays = (pd.to_datetime(optiondata['dtExpiry'].ix[i])-pd.to_datetime(optiondata['dtTrade'].ix[i])).days
        T[i]  = ndays/360
        
    market_price = np.zeros(N)
    for i in range(0,N):
        market_price[i] = bs.vanilla_price(S[i], K[i], r[i]-q[i], market_vol[i], T[i], 'C')
    return market_price
```

# <font face="sans-serif"> Calibration with fixed expiry date

```python
# select all data with the same expiry date

testdata = mydata.loc[mydata['dtExpiry']=='2013-11-21'] 
testdata
```

```python
data=testdata.reset_index() # reset the index
data
```

```python
# filter out useless columns

input_data = DataFrame.filter(data, items=['dtTrade', 'dtExpiry', 'Type','iRate', 'dRate', 'Spot', 'Strike', 'IVMid'])
```

```python
input_data
```

```python
output = heston_calibration(input_data)
```

```python
output
```

```python
mp=market_price(input_data) # market price
```

```python
fp = output[1] # fitted price
```

```python
plot(input_data['Strike'], fp, 'b-', input_data['Strike'], mp, 'ro')
```

# <font face="sans-serif"> Calibration across all expiry dates

```python
start = time.time()
output2 = heston_calibration(mydata)
end = time.time()
print (end-start)
```

```python
output2
```

```python
# append fitted/model price (calculated using the calibrated parameter) to the data

mydata['Fitted Price'] = output2[1]
```

```python
# check that there is now a column for fitted price
mydata
```

```python
mp2=market_price(mydata)
```

```python
mydata['Market Price'] = mp2
```
