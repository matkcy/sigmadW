# Kalman Filter as Linear Regression

## 1. Observation Model

<font face="sans-serif" size='3'>
Suppose we have a series of observations (measurements) $y_{1}$, $y_{1}$, $\ldots$, $y_{r}$ which are meant to tell us something about the state of (the hidden variables) $b_{1}$, $\ldots$, $b_{s}$. Since $y_{i}$'s are measurement, it is reasonable to assume there are errors. We can assume the following observation model:


$$  {\bf y}_{t} = {\bf H} {\bf b}_{t} + {\bf e}_{t}$$




where
$$ {\bf y}_{t} = \begin{pmatrix} y_{1} \\
\vdots \\
y_{r}
						\end{pmatrix}, ~~ {\bf b}_{t} = \begin{pmatrix} b_{1} \\
\vdots \\
b_{s}
\end{pmatrix}, ~~{\bf H}~~\text{is the observation matrix},$$

$$ ~~{\bf e}_{t} ~~\text{is a vector of Gaussian noise}.$$

**Remark.**


1. It is not a typo that the subscript in ${\bf b}_{t}$ is $t$ while the vector has only $s$ rows. This is because the subscript $t$ is used to denote the state variable at time $t$, which is a vector with $s$ components. The same is true for ${\bf y}_{t}$ which has $r$ entries. Normally, when ${\bf y}_{t}$ is some time series, we set $t=r$.

2. In practice, the error ${\bf e}_{t}$ can be determined by specifying its covariance. In fact, we assume ${\bf e}_{t} \sim \text{Normal}(0, {\bf R})$, where ${\bf R}$ is the covariance matrix.

3. In the pyrhon package *pykalman*, the matrix ${\bf H}$ is given by the *observation\_matrices*.

## 2. State (Transition) Model

<font face="sans-serif" size='3'>
We still need a (recursive) state model to describe how the state depends on the previous state:

$${\bf b}_{t} = {\bf A} {\bf b}_{t-1} + {\bf B}{\bf u}_{t-1} + {\bf w}_{t-1},$$


<br>
where ${\bf w}_{t-1}$ is another vector of Gaussian noise, ${\bf u}_{k}$ is some control variable.




The noise  ${\bf w}_{t}$ can be chosen by specifying the state transition covariance matrix ${\bf Q}$. In fact, we assume ${\bf w}_{t} \sim \text{Normal}(0, {\bf Q})$. In the python package *pykalman*, the matrix ${\bf Q}$ is given by the *transition\_covariance*, whereas the matrix ${\bf A}$ is given by the *transition\_matrices*.



## 3. Kalman Filter Recursion

<font face="sans-serif" size='3'>
The core of the Kalman filter is the following recursion. At each iteration, we have the estimate  $(\widehat{{\bf b}}_{t}, {\bf P}_{t})$, where ${\bf P}_{t}$ is the covariance prediction. This is fed into the next iteration to generate the next estimate $(\widehat{{\bf b}}_{t+1},{\bf P}_{t+1})$



Basically, at each iteration, there are two stages in the update:

<br>


1. Time update: where the the state and the covariance prediction are estimated based on their most recent estimations.


\begin{eqnarray*}
\widehat{{\bf b}}_{t}' & = & {\bf A}\widehat{{\bf b}}_{t-1} + {\bf B}{\bf u}_{t}  \\
{\bf P}_{t}' & = & {\bf A} {\bf P}_{t-1} {\bf A}^{T} + {\bf Q}.
\end{eqnarray*}



2. Measurement update: where the Kalman gain ${\bf K}_{t}$ is calculated, and is then used to give a better estimate of the state and covariance prediction. 


\begin{eqnarray*}
{\bf K}_{t} & = & {\bf P}_{t}'{\bf H}^{T} \left({\bf H}{\bf P}_{t}'{\bf H}^{T} + {\bf R} \right)^{-1} \\ 
\widehat{{\bf b}}_{t} & = & \widehat{{\bf b}}_{t}' + {\bf K}_{t} \left({\bf y}_{t} - {\bf H} \widehat{{\bf b}}_{t}'\right)  \\
{\bf P}_{t} & = & \left(1-{\bf K}_{t} {\bf H} \right) {\bf P}_{t}'
\end{eqnarray*}



**Remarks.**

1. The matrix ${\bf R}$ is related to the measurement noise ${\bf e}_{t}$. In the python package *pykalman*, this is refered to as the *observation\_ covariance*. 

2. The equation $\widehat{{\bf b}}_{t}  = \widehat{{\bf b}}_{t}' + {\bf K}_{t} \left({\bf y}_{t} - {\bf H} \widehat{{\bf b}}_{t}'\right)$ roughly says that the current estimate is the combination of the current time update (based on the state model) and that of Kalman gain multiplied by the measurement error.

3. To start off the recursion, we need the initial estimate  $(\widehat{{\bf b}}_{0}, {\bf P}_{0})$. In the python package *pykalman*, these are given by *initial\_state\_mean* and *initial\_state\_covariance* respectively.



## 4. An Example: Online Linear Regression

<font face="sans-serif" size='3'>
Suppose we want to find the hedge ratio between two price series $y_{1}$, $\ldots$, $y_{t}$ and $x_{1}$, $\ldots$, $x_{t}$:

$$ y_{i} = \beta_{t}  x_{i} + c_{t}, ~~1 \le i \le t,$$

<br>

where $\beta_{t}$ is the hedge ratio and $c_{t}$ is the intercept. We can write this as

$$ {\bf y}_{t} = {\bf H} {\bf b}_{t} + {\bf e}_{t}$$

where

$$ {\bf y}_{t} = \begin{pmatrix} y_{1}\\
\vdots \\
y_{t}
						\end{pmatrix}, ~~{\bf H} = \begin{pmatrix} x_{1} & 1 \\
\vdots & \vdots \\
x_{t} & 1 
\end{pmatrix}, ~~{\bf b}_{t} = \begin{pmatrix} \beta_{t}\\
c_{t} \end{pmatrix}.$$



For simplicity, with the benefit of hindsight, we assume the noises compnents are independent and each has variance $0.01$.

**Idea:** Instead of regressing the entire series to get one linear fit, we want to update $\beta_{t}$ and $c_{t}$ based on the price series as well as the most recent estimate $\beta_{t-1}$ and $c_{t-1}$. That is, we want to update the hedge ratio and the intercept as $t$ advances such that  in the long run, we expect the estimates to be less dependent on past states. One can achieve this by a fixed weighting scheme, however, it is not clear whether the weights assigned are optimal. Kalman filter tries to find the most optimum averaging factor for each consequent state, and somehow remembers a little bit about the past states on the fly!




For simplicity, we choose the following state model:

$$\begin{pmatrix} \beta_{t} \\
c_{t} \end{pmatrix} = \begin{pmatrix} \beta_{t-1} \\ c_{t-1} \end{pmatrix} + {\bf w}_{t-1}.$$

In this case, ${\bf A} = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}$, ${\bf B} = {\bf 0}$. Also, we are going to choose


$$ {\bf Q} = \frac{\delta}{1-\delta} \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}, ~~\text{where}~~\delta = 0.00001.$$


```python
import pandas as pd
import os, os.path
import datetime as dt
from pandas.io.data import DataReader
from pykalman import KalmanFilter
csv_dir = '/Users/matkcy/Documents/MyPython/algo/all_data/'
```

```python
# get prices from local csv file
data={}
pair = ['EWA', 'EWC']
start = dt.datetime(2006,4,26)
end = dt.datetime(2015,1,1)
#data = DataReader(pair, 'yahoo', start, end)['Adj Close']
for s in pair:
    data[s] = pd.io.parsers.read_csv(
                os.path.join(csv_dir, '%s.csv' % s),
                header=0, index_col=0, parse_dates=True,
                names=[
                    'datetime', 'open', 'high', 
                    'low', 'close', 'volume', 'adj_close'
                ]
            ).sort()
```

```python
# get adjusted close price for two assets
for s in pair:
    data[s]=data[s][start:end]
x = data['EWA']['adj_close'].values
y = data['EWC']['adj_close'].values
```

```python
kf = KalmanFilter(#n_dim_obs=1, n_dim_state=2,
                  initial_state_mean=np.zeros(2),
                  initial_state_covariance=np.ones((2, 2)),
                  transition_matrices=np.eye(2),
                  observation_matrices=obs_mat,
                  observation_covariance=0.01,
                  transition_covariance=trans_cov)
```

```python
#state_means, state_covs = kf.filter(y)
state_means, state_covs = kf.filter(y)
```

```python
pd.DataFrame(dict(slope=state_means[:, 0], intercept=state_means[:, 1]), index=I).plot(subplots=True)
plt.tight_layout()

#plt.savefig('slope_intercept.png')
```
