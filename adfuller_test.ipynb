## <font face="sans-serif"> Ornstein-Uhlenbeck Process as mean-reverting process</font>

<font face="sans-serif" size='3'>

<p>
The Ornstein-Uhlenbeck process is the stochastic process represented by the differential equation:
</p>

<p>
$$ dV_{t} = \theta(\mu - V_{t}) \, dt + \sigma dW_{t},$$
</p>

<p>
where $\theta<0$, $\mu$ and $\sigma >0$ are parameters.
</p>



<p> This process is stationary (a stochastic process whose joint probability distribution does not change when shifted in time. Consequently, parameters such as the mean and variance, if they are present, also do not change over time and do not follow any trends.), Gaussian, and Markovian, and is the only nontrivial process that satisfies these three conditions, up to allowing linear transformations of the space and time variables. Over time, the process tends to drift towards its long-term mean: such a process is called mean-reverting.


</font>

## <font face="sans-serif"> Augmented Dickey-Fuller (ADF) Test </font>

<font face="sans-serif" size='3'>


Suppose we have a time series $(y_{t})$.  Consider a linear lag model of order $p$ of this time series which postulates that

<br>
<br>



$$ \triangle y_{t} = a + bt + c y_{t-1} + d_{1} \triangle y_{t-1} + \cdots + d_{p-1} \triangle y_{t-p+1} + \varepsilon_{t}.$$




<br>


Here, $a$, $b$ and $c$ are constants and $\triangle y_{t} = y_{t}-y_{t-1}$, and $\varepsilon_{t}$ is a noise term. The term $a$ is the intercept term. When the time series is 
flat (no trend) and potentially slow-turning around a
non-zero value, $a$ is not zero. The term $b$ is the time trend. When the time series 
has a trend in it (either up or down) and is potentially slow-turning around a trend line, $b$ is not zero.

<br>
<br>



For our purpose, we are going to assume that $b=d_{1} = \cdots = d_{p-1} = 0$. This is motivated by the Ornstein-Uhlenbeck process (see above).

<br>
<br>

Notice that if $c<0$, then the change in $y_{t}$ is proportional to the current value $y_{t-1}$ and thus it is likely to be mean-reverting.


<br>
<br>


The ADF Test is carried out under the  null hypothesis $c = 0$ (non-stationary) against the alternative hypothesis $c<0$ (stationary). The test is carried out as follows:



<br>
<br>

<li> The test statistic $\tau$ is calculated as follows:
$$\tau = \frac{\hat{c}}{SE(\hat{c})}.$$

<br>
<br>

<li> The test statistic $\tau$ is then compared to the relevant critical values (at $1\%$, $5\%$, $10\%$) 


<br>
<br>

<li> If $\tau < p$ where $p$ is the critical value at $x\%$, then we reject the hypothesis at $(100-x)\%$ confidence, i.e. the time series is mean-reverting.


<br>
<br>

<li> If $\tau$ is more than all the critical values, then the time series is unlikely to be mean-reverting.


</font>

## <font face="sans-serif"> Testing for mean-reversion for all series in the list </font>

```python
def ADF_List(symbol_list, start, end):
    for key in symbol_list.keys():
        currency_pair = symbol_list[key]
        df = web.DataReader(currency_pair, 'fred', start, end)
        T=df[currency_pair].notnull()
        for date in df.index:
            if T.loc[date]==False:
                df = df.drop(date)
        df = df.sort_index()

        result = ts.adfuller(df[currency_pair],1)
        test_statistic = result[0]
        critical_value_1 = result[4]['1%']
        critical_value_5 = result[4]['5%']
        critical_value_10 = result[4]['10%']
        print ('ADF Test for'), (currency_pair)
        if test_statistic<critical_value_1:
            print ('99%: *** reject ***')
        else:
            print ('99%: cannot reject')
        if test_statistic<critical_value_5:
            print ('95%: *** reject ***')
        else:
            print ('95%: cannot reject')
        if test_statistic<critical_value_10:
            print ('90%: *** reject ***')
        else:
            print ('90%: cannot reject')
    
    print ('...Done')
    
```

```python
start = dt.datetime(2007,1,1)
end = dt.datetime(2008,1,1)
ADF_List(symbol_list, start, end)
```
