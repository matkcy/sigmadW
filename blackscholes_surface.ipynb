# Black-Scholes Surface

<font face="sans-serif" size='3'>
In this notebook, we show how to plot the Black-Scholes surface for European call option prices. The pricing engine The <a href=blackscholespricer.py>blackscholespricer</a> module contains the pricing under the usual Black-Scholes model.

```python
import blackscholespricer as bs
import numpy as np
```

```python
'''option's parameters:
K=strike, r=interest rate, sigma=volatility, T=time to maturity, option_type = call/put'''

K=100 
r=0.1 
sigma=0.3 
T=1 
option_type = 'C'
```

```python
# A single call price 
bs.vanilla_price(100, K, r, sigma, T, option_type)
```

```python
# Vectorize the fucntion bs.vanilla_price
call_price = np.vectorize(bs.vanilla_price)
```

```python
# max stock price
Smax = 200
```

```python
Svec  = np.linspace(0, Smax, 100)
Tvec = np.linspace(0, T, 100)
```

```python
X,Y = np.meshgrid(Svec,Tvec)
Z = np.zeros(shape=X.shape)
```

```python
Z = call_price(X,K,r,sigma,Y,'C')
```
